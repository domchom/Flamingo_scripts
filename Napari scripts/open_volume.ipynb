```python
import os
import numpy as np
from pathlib import Path
from tifffile import imread
import napari
```

```python
fpath = Path('/Volumes/zs2tb/20220318_150236_E02Series/0_vol_downsample')
files = [file for file in os.listdir(fpath) if file.endswith('.tif') and not file.startswith('.')]
chan1_files = [file for file in files if 'C00' in file]
chan2_files = [file for file in files if 'C02' in file]
chn1 = np.zeros(shape=(366, 48, 256, 256))
chn2 = np.zeros(shape=(366, 48, 256, 256))

for i in range(366):
    chn1[i] = imread(fpath / chan1_files[i])
    chn2[i] = imread(fpath / chan2_files[i])
    print(f'finished loading time point {i}')

print('done loading arrays')
```

```python
%gui qt
```

```python
viewer = napari.Viewer()
```

```python
viewer.add_image(chn1, colormap='green', blending='additive', scale=(1.75,1,1))
viewer.add_image(chn2, colormap='magenta', blending='additive', scale=(1.75,1,1))
```
